```python
#собираем данные по каждому региону (по ТИКам)
def get_region_data(link,data_count):
    req = requests.get(link)
    result = BeautifulSoup(req.text, 'lxml')
    TIKs = result.findAll("table")[1].findAll("table")[7].find("tr").findAll("td")
    tiks = []
    PATH_FILE='data'
    if(len(TIKs)==0):
        data_link = result.find("a")['href']
        req = requests.get(data_link)
        file = open(PATH_FILE + "\data{}.html".format(data_count),"w",encoding='utf-8');
        file.write(req.text);
        file.close();
        data_count+=1;
        
    for tik in TIKs:
        tiks_name = tik.find("a").text
        href = tik.find("a")['href']
        print(data_count)
        request = requests.get(href)
        uik_link = BeautifulSoup(request.text, 'lxml').findAll("table")[1].findAll("tr")[1].find("td").find("a")['href']
        request = requests.get(uik_link)        
        file = open(PATH_FILE + "\data{}.html".format(data_count),"w",encoding='utf-8');
        file.write(request.text);
        file.close();
        data_count+=1;
    return data_count
```

```python
#собираем страницы по каждому ТИК для каждого региона
regions_data = []
data_count = 0
for region in regions:
    region_name = region.find("a").text
    data_count = get_region_data(region.find("a")['href'], data_count)
    
```

```python
#считываем данные с файлов и заносим в словарь, где ключ - id страницы, значение - массив статистики по УИК
def get_data():
    fds = sorted(os.listdir('data/'))
    tiks = {}
    k=0
    for file in fds:
        print('data/{}'.format(file))
        file_data = open('data/{}'.format(file),"r",encoding='utf-8');
        document = BeautifulSoup(file_data.read())
        tables = document.findAll("table")
        table_count = len(tables)
        
        f = furl(tables[table_count-9].findAll("a")[1]['href'])
        print(f)
        if '{}'.format(file) == 'data2776.html' or '{}'.format(file) == 'data2777.html' :
            f = furl(tables[table_count-9].findAll("a")[0]['href'])
        tiks_id = f.args['tvd']
    
        current_table = tables[table_count-9].findAll("table");
        tiks_name = current_table[2].findAll("tr")[1].findAll("td")[1].text
        statistics_str = current_table[6].findAll('tr') #массив данных
        search =current_table[7];
        uiks = []
        i = 0
        out_data = []
        x = search.findAll("tr")
        uiks = x[0].findAll('td') #массив данных
        j = 0;
        for uik in uiks:
            array = []
            i = 0
            #20 полей для УИК
            for data in statistics_str:
                if i == 0 or i == 13:
                    i+=1
                    continue
                #tds = data.findAll("td")
                count = x[i].findAll("td")[j].find("nobr").find("b").text
                array.append(count)
                i+=1
            uik_name = uik.find("nobr").text
            out_data.append([uik_name, array])
            j+=1
        tiks[tiks_id] = out_data
        print(k)
        k+=1
    #print(tiks)
    return tiks
```

```python
#формируем финальную структуру данных:
# [Название_региона: [Название_ТИК: [Номер_УИК:[Статистические_данные]]]]
regions_data = []
data_count = 0
for region in regions:
    region_name = region.find("a").text
    req = requests.get(region.find("a")['href'])
    res = BeautifulSoup(req.text, 'lxml')
    TIKs = res.findAll("table")[1].findAll("table")[7].find("tr").findAll("td")
    data = []
    #print(len(TIKs))
    if(len(TIKs) == 0):
        tiks_name = None
        href = res.findAll("table")[1].findAll("tr")[1].find("a")['href']
        f = furl(href)
        print(f)
        tik_id = f.args['vibid']
        data.append([tiks_name,result.get(tik_id)])
    for tik in TIKs:
        tiks_name = tik.find("a").text  
        href = tik.find("a")['href']
        f = furl(href)
        tik_id = f.args['vibid']
        data.append([tiks_name,result.get(tik_id)])
    regions_data.append([region_name, data])
    print(data_count)
    data_count+=1
    
```
